```python
import tensorflow.keras.preprocessing.image as image
from tensorflow.keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D
from tensorflow.keras.models import Model
```

```python
%matplotlib inline
img = image.load_img('Images Night\\frame0.jpg',grayscale=True,target_size = (100,100))
img_array = image.img_to_array(img)
print(img_array.shape)
img.show()
```

```python
input_img = Input(shape=(100, 100, 1))
conv2d_1 = Conv2D(4, (3, 3), activation='relu', padding='same')(input_img)
encoded = conv2d_1
conv2d_2 = Conv2D(16, (3, 3), activation='relu', padding='same')(conv2d_1)
decoded = conv2d_1
autoencoder = Model(input_img, decoded)
autoencoder.compile(optimizer='adadelta',loss="mean_absolute_error")
```

```python
from os import listdir
loaded_images = list()
i=0
for filename in listdir('Images Night'):
    # load image
    img = image.load_img('Images Night\\' + filename, grayscale = True, target_size = (100,100))
    # store loaded image
    loaded_images.append(img)
    i=i+1
    print('> loaded %d %s %s' % (i,filename, img.size))
```

```python
%matplotlib inline
from matplotlib.pyplot import figure
figure(num=None, figsize=(5, 100), dpi=80, facecolor='w', edgecolor='k')

for layer in autoencoder.layers:
    # check for convolutional layer
    if 'conv' not in layer.name:
        continue
    # get filter weights
    filters, biases = layer.get_weights()
    print(layer.name, filters.shape)

print(filters.shape[3])
# plot first few filters
n_filters, ix = filters.shape[3], 1
for i in range(n_filters):
    # get the filter
    f = filters[:, :, :, i]
    # specify subplot and turn of axis
    ax = plt.subplot(30, 3, ix)
    ax.set_xticks([])
    ax.set_yticks([])
    # plot filter channel in grayscale
    plt.imshow(f[:,:,0], cmap='gray')
    ix += 1
# show the figure
plt.show()
```
